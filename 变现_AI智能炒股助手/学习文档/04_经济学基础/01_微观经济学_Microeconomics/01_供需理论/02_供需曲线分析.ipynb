## 3. å¸‚åœºå‡è¡¡


```python
# è®¾ç½®å›¾å½¢å¤§å°
plt.figure(figsize=(12, 8))

# éœ€æ±‚æ›²çº¿ï¼šQd = 100 - 2P
# ä¾›ç»™æ›²çº¿ï¼šQs = 20 + 2P
prices = np.linspace(0, 50, 100)
demand = 100 - 2 * prices
supply = 20 + 2 * prices

# è®¡ç®—å‡è¡¡ç‚¹ï¼šQd = Qs
# 100 - 2P = 20 + 2P
# 80 = 4P
# P = 20
equilibrium_price = 20
equilibrium_quantity = 100 - 2 * equilibrium_price  # æˆ– 20 + 2 * 20 = 60

plt.plot(demand, prices, 'b-', linewidth=2, label='éœ€æ±‚æ›²çº¿ D')
plt.plot(supply, prices, 'r-', linewidth=2, label='ä¾›ç»™æ›²çº¿ S')

# æ ‡æ³¨å‡è¡¡ç‚¹
plt.plot(equilibrium_quantity, equilibrium_price, 'go', markersize=12, label='å‡è¡¡ç‚¹ E')
plt.axvline(x=equilibrium_quantity, color='gray', linestyle='--', alpha=0.5)
plt.axhline(y=equilibrium_price, color='gray', linestyle='--', alpha=0.5)

plt.xlabel('æ•°é‡ (Q)', fontsize=12)
plt.ylabel('ä»·æ ¼ (P)', fontsize=12)
plt.title('å¸‚åœºå‡è¡¡', fontsize=14, fontweight='bold')
plt.grid(True, alpha=0.3)
plt.legend(fontsize=11, loc='upper right')
plt.xlim(0, 120)
plt.ylim(0, 55)

# æ ‡æ³¨å‡è¡¡ä»·æ ¼å’Œæ•°é‡
plt.text(equilibrium_quantity + 5, equilibrium_price, 
         f'E({equilibrium_quantity}, {equilibrium_price})',
         fontsize=11, fontweight='bold')

plt.tight_layout()
plt.show()

print(f"å‡è¡¡ä»·æ ¼: {equilibrium_price}")
print(f"å‡è¡¡æ•°é‡: {equilibrium_quantity}")
```

## 4. éœ€æ±‚å˜åŠ¨ï¼ˆéœ€æ±‚æ›²çº¿ç§»åŠ¨ï¼‰


```python
# è®¾ç½®å›¾å½¢å¤§å°
plt.figure(figsize=(12, 8))

prices = np.linspace(0, 50, 100)

# åŸå§‹éœ€æ±‚æ›²çº¿ï¼šQd = 100 - 2P
demand_original = 100 - 2 * prices

# éœ€æ±‚å¢åŠ ï¼šQd = 120 - 2Pï¼ˆéœ€æ±‚æ›²çº¿å³ç§»ï¼‰
demand_increase = 120 - 2 * prices

# éœ€æ±‚å‡å°‘ï¼šQd = 80 - 2Pï¼ˆéœ€æ±‚æ›²çº¿å·¦ç§»ï¼‰
demand_decrease = 80 - 2 * prices

# ä¾›ç»™æ›²çº¿ï¼šQs = 20 + 2P
supply = 20 + 2 * prices

plt.plot(demand_original, prices, 'b-', linewidth=2, label='åŸå§‹éœ€æ±‚æ›²çº¿ D0')
plt.plot(demand_increase, prices, 'g--', linewidth=2, label='éœ€æ±‚å¢åŠ  D1')
plt.plot(demand_decrease, prices, 'r--', linewidth=2, label='éœ€æ±‚å‡å°‘ D2')
plt.plot(supply, prices, 'k-', linewidth=2, alpha=0.5, label='ä¾›ç»™æ›²çº¿ S')

# è®¡ç®—æ–°çš„å‡è¡¡ç‚¹
# éœ€æ±‚å¢åŠ ï¼š120 - 2P = 20 + 2P, P = 25, Q = 70
# éœ€æ±‚å‡å°‘ï¼š80 - 2P = 20 + 2P, P = 15, Q = 50
plt.plot(60, 20, 'bo', markersize=10, label='å‡è¡¡ç‚¹ E0')
plt.plot(70, 25, 'go', markersize=10, label='å‡è¡¡ç‚¹ E1')
plt.plot(50, 15, 'ro', markersize=10, label='å‡è¡¡ç‚¹ E2')

plt.xlabel('æ•°é‡ (Q)', fontsize=12)
plt.ylabel('ä»·æ ¼ (P)', fontsize=12)
plt.title('éœ€æ±‚å˜åŠ¨å¯¹å‡è¡¡çš„å½±å“', fontsize=14, fontweight='bold')
plt.grid(True, alpha=0.3)
plt.legend(fontsize=10, loc='upper right')
plt.xlim(0, 130)
plt.ylim(0, 55)

plt.tight_layout()
plt.show()

print("éœ€æ±‚å˜åŠ¨çš„å½±å“ï¼š")
print("éœ€æ±‚å¢åŠ  â†’ å‡è¡¡ä»·æ ¼ä¸Šå‡ï¼Œå‡è¡¡æ•°é‡å¢åŠ ")
print("éœ€æ±‚å‡å°‘ â†’ å‡è¡¡ä»·æ ¼ä¸‹é™ï¼Œå‡è¡¡æ•°é‡å‡å°‘")
```

## 5. ä¾›ç»™å˜åŠ¨ï¼ˆä¾›ç»™æ›²çº¿ç§»åŠ¨ï¼‰


```python
# è®¾ç½®å›¾å½¢å¤§å°
plt.figure(figsize=(12, 8))

prices = np.linspace(0, 50, 100)

# éœ€æ±‚æ›²çº¿ï¼šQd = 100 - 2P
demand = 100 - 2 * prices

# åŸå§‹ä¾›ç»™æ›²çº¿ï¼šQs = 20 + 2P
supply_original = 20 + 2 * prices

# ä¾›ç»™å¢åŠ ï¼šQs = 40 + 2Pï¼ˆä¾›ç»™æ›²çº¿å³ç§»ï¼‰
supply_increase = 40 + 2 * prices

# ä¾›ç»™å‡å°‘ï¼šQs = 2Pï¼ˆä¾›ç»™æ›²çº¿å·¦ç§»ï¼‰
supply_decrease = 2 * prices

plt.plot(demand, prices, 'k-', linewidth=2, alpha=0.5, label='éœ€æ±‚æ›²çº¿ D')
plt.plot(supply_original, prices, 'b-', linewidth=2, label='åŸå§‹ä¾›ç»™æ›²çº¿ S0')
plt.plot(supply_increase, prices, 'g--', linewidth=2, label='ä¾›ç»™å¢åŠ  S1')
plt.plot(supply_decrease, prices, 'r--', linewidth=2, label='ä¾›ç»™å‡å°‘ S2')

# è®¡ç®—æ–°çš„å‡è¡¡ç‚¹
# ä¾›ç»™å¢åŠ ï¼š100 - 2P = 40 + 2P, P = 15, Q = 70
# ä¾›ç»™å‡å°‘ï¼š100 - 2P = 2P, P = 25, Q = 50
plt.plot(60, 20, 'bo', markersize=10, label='å‡è¡¡ç‚¹ E0')
plt.plot(70, 15, 'go', markersize=10, label='å‡è¡¡ç‚¹ E1')
plt.plot(50, 25, 'ro', markersize=10, label='å‡è¡¡ç‚¹ E2')

plt.xlabel('æ•°é‡ (Q)', fontsize=12)
plt.ylabel('ä»·æ ¼ (P)', fontsize=12)
plt.title('ä¾›ç»™å˜åŠ¨å¯¹å‡è¡¡çš„å½±å“', fontsize=14, fontweight='bold')
plt.grid(True, alpha=0.3)
plt.legend(fontsize=10, loc='upper right')
plt.xlim(0, 130)
plt.ylim(0, 55)

plt.tight_layout()
plt.show()

print("ä¾›ç»™å˜åŠ¨çš„å½±å“ï¼š")
print("ä¾›ç»™å¢åŠ  â†’ å‡è¡¡ä»·æ ¼ä¸‹é™ï¼Œå‡è¡¡æ•°é‡å¢åŠ ")
print("ä¾›ç»™å‡å°‘ â†’ å‡è¡¡ä»·æ ¼ä¸Šå‡ï¼Œå‡è¡¡æ•°é‡å‡å°‘")
```

## 6. è‚¡ç¥¨å¸‚åœºä¾›éœ€åˆ†æç¤ºä¾‹


```python
# æ¨¡æ‹Ÿè‚¡ç¥¨å¸‚åœºçš„ä¾›éœ€å…³ç³»
# å‡è®¾æŸè‚¡ç¥¨çš„éœ€æ±‚å’Œä¾›ç»™å‡½æ•°

prices = np.linspace(10, 50, 100)

# éœ€æ±‚æ›²çº¿ï¼šæŠ•èµ„è€…æ„¿æ„ä¹°å…¥çš„æ•°é‡
# Qd = 1000 - 15Pï¼ˆä»·æ ¼è¶Šé«˜ï¼Œä¹°å…¥æ„æ„¿è¶Šä½ï¼‰
demand = 1000 - 15 * prices
demand = np.maximum(demand, 0)  # ç¡®ä¿éè´Ÿ

# ä¾›ç»™æ›²çº¿ï¼šè‚¡ä¸œæ„¿æ„å–å‡ºçš„æ•°é‡
# Qs = 200 + 10Pï¼ˆä»·æ ¼è¶Šé«˜ï¼Œå–å‡ºæ„æ„¿è¶Šé«˜ï¼‰
supply = 200 + 10 * prices

# è®¡ç®—å‡è¡¡ç‚¹
# 1000 - 15P = 200 + 10P
# 800 = 25P
# P = 32
equilibrium_price = 32
equilibrium_quantity = 1000 - 15 * equilibrium_price

plt.figure(figsize=(12, 8))
plt.plot(demand, prices, 'b-', linewidth=2, label='éœ€æ±‚æ›²çº¿ï¼ˆä¹°å…¥æ„æ„¿ï¼‰')
plt.plot(supply, prices, 'r-', linewidth=2, label='ä¾›ç»™æ›²çº¿ï¼ˆå–å‡ºæ„æ„¿ï¼‰')
plt.plot(equilibrium_quantity, equilibrium_price, 'go', markersize=12, label='å‡è¡¡ç‚¹')
plt.axvline(x=equilibrium_quantity, color='gray', linestyle='--', alpha=0.5)
plt.axhline(y=equilibrium_price, color='gray', linestyle='--', alpha=0.5)

plt.xlabel('äº¤æ˜“é‡ (ä¸‡è‚¡)', fontsize=12)
plt.ylabel('è‚¡ä»· (å…ƒ)', fontsize=12)
plt.title('è‚¡ç¥¨å¸‚åœºä¾›éœ€åˆ†æ', fontsize=14, fontweight='bold')
plt.grid(True, alpha=0.3)
plt.legend(fontsize=11)
plt.xlim(0, 1200)
plt.ylim(0, 55)

plt.text(equilibrium_quantity + 50, equilibrium_price, 
         f'å‡è¡¡ä»·æ ¼: {equilibrium_price}å…ƒ\\nå‡è¡¡æ•°é‡: {equilibrium_quantity:.0f}ä¸‡è‚¡',
         fontsize=11, bbox=dict(boxstyle='round', facecolor='wheat', alpha=0.5))

plt.tight_layout()
plt.show()

print(f"è‚¡ç¥¨å¸‚åœºå‡è¡¡åˆ†æï¼š")
print(f"å‡è¡¡ä»·æ ¼: {equilibrium_price} å…ƒ")
print(f"å‡è¡¡äº¤æ˜“é‡: {equilibrium_quantity:.0f} ä¸‡è‚¡")
```
