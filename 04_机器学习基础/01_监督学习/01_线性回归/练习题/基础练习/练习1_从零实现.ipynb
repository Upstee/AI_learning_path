### 2. 核心算法

**梯度下降更新规则**：
- 计算预测值：$h = Xw + b$
- 计算损失：$J = \frac{1}{2m}\sum(h - y)^2$
- 计算梯度：
  - $\frac{\partial J}{\partial w} = \frac{1}{m}X^T(h - y)$
  - $\frac{\partial J}{\partial b} = \frac{1}{m}\sum(h - y)$
- 更新参数：
  - $w := w - \alpha \frac{\partial J}{\partial w}$
  - $b := b - \alpha \frac{\partial J}{\partial b}$
