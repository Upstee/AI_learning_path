## 1. åŠ¨ä½œåºåˆ—çš„åŸºæœ¬åŸç†

### 1.1 ä»€ä¹ˆæ˜¯åŠ¨ä½œåºåˆ—

**ç›´è§‚ç†è§£**ï¼šæƒ³è±¡åŠ¨ä½œåºåˆ—å°±åƒæ‰§è¡Œä»»åŠ¡çš„æ­¥éª¤æ¸…å•ï¼Œæ¯ä¸ªæ­¥éª¤æŒ‰é¡ºåºæ‰§è¡Œã€‚

**å®šä¹‰**ï¼šåŠ¨ä½œåºåˆ—æ˜¯æŒ‡æŒ‰æ—¶é—´é¡ºåºæ’åˆ—çš„åŠ¨ä½œåºåˆ—ï¼Œæ˜¯åŠ¨ä½œè¡¨ç¤ºçš„ä¸€ç§æ–¹æ³•ã€‚

### 1.2 åŠ¨ä½œåºåˆ—çš„æ•°å­¦è¡¨ç¤º

åŠ¨ä½œåºåˆ—çš„æ•°å­¦è¡¨ç¤ºä¸ºï¼š

$$A = [a_1, a_2, \ldots, a_T]$$

å…¶ä¸­ï¼š
- $A$ æ˜¯åŠ¨ä½œåºåˆ—
- $a_t$ æ˜¯ç¬¬$t$ä¸ªæ—¶é—´æ­¥çš„åŠ¨ä½œ
- $T$ æ˜¯åºåˆ—é•¿åº¦

æ¯ä¸ªåŠ¨ä½œ$a_t$å¯ä»¥æ˜¯ç¦»æ•£åŠ¨ä½œæˆ–è¿ç»­åŠ¨ä½œï¼š
- **ç¦»æ•£åŠ¨ä½œ**ï¼š$a_t \in \{0, 1, \ldots, |\mathcal{A}|-1\}$
- **è¿ç»­åŠ¨ä½œ**ï¼š$a_t \in \mathbb{R}^d$

### 1.3 æ—¶åºä¿¡æ¯

åŠ¨ä½œåºåˆ—ä¸­çš„æ—¶åºä¿¡æ¯åŒ…æ‹¬ï¼š

1. **æ—¶é—´é¡ºåº**ï¼šåŠ¨ä½œçš„æ‰§è¡Œé¡ºåº
2. **åŠ¨ä½œæŒç»­æ—¶é—´**ï¼šæ¯ä¸ªåŠ¨ä½œçš„æŒç»­æ—¶é—´
3. **åŠ¨ä½œé—´éš”**ï¼šåŠ¨ä½œä¹‹é—´çš„é—´éš”æ—¶é—´

### 1.4 åŠ¨ä½œåºåˆ—çš„å¯è§†åŒ–

ä¸‹é¢æˆ‘ä»¬é€šè¿‡ä»£ç å¯è§†åŒ–åŠ¨ä½œåºåˆ—çš„ç»“æ„ï¼š


````python
# ============================================
# åŠ¨ä½œåºåˆ—å¯è§†åŒ–ï¼ˆç¤ºä¾‹ï¼šVLAä»»åŠ¡çš„åŠ¨ä½œåºåˆ—ï¼‰
# ============================================
np.random.seed(42)

# ç¤ºä¾‹ï¼šVLAä»»åŠ¡"æ‹¿èµ·æ¯å­æ”¾åˆ°æ¡Œå­ä¸Š"çš„åŠ¨ä½œåºåˆ—
task_name = "æ‹¿èµ·æ¯å­æ”¾åˆ°æ¡Œå­ä¸Š"
action_sequence = [
    {'name': 'ç§»åŠ¨åˆ°æ¯å­ä½ç½®', 'time': 0.0, 'duration': 2.0, 'type': 'ç§»åŠ¨'},
    {'name': 'æŠ“å–æ¯å­', 'time': 2.0, 'duration': 1.0, 'type': 'æŠ“å–'},
    {'name': 'ç§»åŠ¨åˆ°æ¡Œå­ä½ç½®', 'time': 3.0, 'duration': 2.0, 'type': 'ç§»åŠ¨'},
    {'name': 'æ”¾ä¸‹æ¯å­', 'time': 5.0, 'duration': 1.0, 'type': 'æ”¾ç½®'},
]
sequence_length = len(action_sequence)

# æå–æ—¶é—´ä¿¡æ¯
action_times = [a['time'] for a in action_sequence]
action_durations = [a['duration'] for a in action_sequence]
action_names = [a['name'] for a in action_sequence]
action_types = [a['type'] for a in action_sequence]

# å¯è§†åŒ–
fig, axes = plt.subplots(2, 2, figsize=(16, 12))

# å·¦ä¸Šï¼šåŠ¨ä½œåºåˆ—æ—¶é—´è½´
ax1 = axes[0, 0]
colors_map = {'ç§»åŠ¨': 'blue', 'æŠ“å–': 'red', 'æ”¾ç½®': 'green'}
y_pos = np.arange(sequence_length)
for i, (name, time, duration, action_type) in enumerate(zip(action_names, action_times, action_durations, action_types)):
    color = colors_map.get(action_type, 'gray')
    ax1.barh(i, duration, left=time, color=color, alpha=0.7, edgecolor='black', linewidth=2)
    ax1.text(time + duration/2, i, name, ha='center', va='center', 
            fontsize=10, fontweight='bold', color='white')

ax1.set_yticks(y_pos)
ax1.set_yticklabels([f'æ­¥éª¤{i+1}' for i in range(sequence_length)])
ax1.set_xlabel('æ—¶é—´ (ç§’)')
ax1.set_ylabel('åŠ¨ä½œæ­¥éª¤')
ax1.set_title(f'åŠ¨ä½œåºåˆ—æ—¶é—´è½´ï¼š{task_name}', fontsize=12, fontweight='bold')
ax1.set_xlim(-0.5, max(action_times) + max(action_durations) + 0.5)
ax1.invert_yaxis()
ax1.grid(True, alpha=0.3, axis='x')

# å³ä¸Šï¼šåŠ¨ä½œæŒç»­æ—¶é—´
ax2 = axes[0, 1]
colors_bar = [colors_map.get(t, 'gray') for t in action_types]
bars = ax2.bar(range(sequence_length), action_durations, color=colors_bar, 
              alpha=0.7, edgecolor='black', linewidth=2)
ax2.set_title('åŠ¨ä½œæŒç»­æ—¶é—´', fontsize=12, fontweight='bold')
ax2.set_xlabel('åŠ¨ä½œæ­¥éª¤')
ax2.set_ylabel('æŒç»­æ—¶é—´ (ç§’)')
ax2.set_xticks(range(sequence_length))
ax2.set_xticklabels([f'æ­¥éª¤{i+1}' for i in range(sequence_length)])
ax2.grid(True, alpha=0.3, axis='y')

# æ·»åŠ æ•°å€¼æ ‡æ³¨
for i, (bar, duration) in enumerate(zip(bars, action_durations)):
    height = bar.get_height()
    ax2.text(bar.get_x() + bar.get_width()/2., height + 0.05,
             f'{duration:.1f}s', ha='center', va='bottom', fontsize=10, fontweight='bold')

# å·¦ä¸‹ï¼šåŠ¨ä½œé—´éš”
ax3 = axes[1, 0]
intervals = [action_times[i+1] - (action_times[i] + action_durations[i]) 
            for i in range(sequence_length - 1)]
intervals = [0] + intervals  # ç¬¬ä¸€ä¸ªåŠ¨ä½œæ²¡æœ‰é—´éš”
bars = ax3.bar(range(sequence_length), intervals, color='orange', 
              alpha=0.7, edgecolor='black', linewidth=2)
ax3.set_title('åŠ¨ä½œé—´éš”æ—¶é—´', fontsize=12, fontweight='bold')
ax3.set_xlabel('åŠ¨ä½œæ­¥éª¤')
ax3.set_ylabel('é—´éš”æ—¶é—´ (ç§’)')
ax3.set_xticks(range(sequence_length))
ax3.set_xticklabels([f'æ­¥éª¤{i+1}' for i in range(sequence_length)])
ax3.grid(True, alpha=0.3, axis='y')

# æ·»åŠ æ•°å€¼æ ‡æ³¨
for i, (bar, interval) in enumerate(zip(bars, intervals)):
    height = bar.get_height()
    if height > 0:
        ax3.text(bar.get_x() + bar.get_width()/2., height + 0.01,
                f'{interval:.1f}s', ha='center', va='bottom', fontsize=10, fontweight='bold')

# å³ä¸‹ï¼šåŠ¨ä½œåºåˆ—æµç¨‹å›¾
ax4 = axes[1, 1]
ax4.axis('off')
y_pos = 0.9
for i, (name, action_type) in enumerate(zip(action_names, action_types)):
    y_pos -= 0.15
    color = colors_map.get(action_type, 'gray')
    ax4.text(0.5, y_pos, f'æ­¥éª¤{i+1}: {name}', ha='center', va='center', 
            fontsize=11, fontweight='bold',
            bbox=dict(boxstyle='round', facecolor=color, alpha=0.7))
    if i < sequence_length - 1:
        ax4.arrow(0.5, y_pos - 0.05, 0, -0.05, head_width=0.03, head_length=0.02, 
                 fc='black', ec='black')

ax4.set_xlim(0, 1)
ax4.set_ylim(0, 1)
ax4.set_title('åŠ¨ä½œåºåˆ—æµç¨‹å›¾', fontsize=12, fontweight='bold')

plt.tight_layout()
plt.suptitle(f'åŠ¨ä½œåºåˆ—å¯è§†åŒ–ï¼š{task_name}', fontsize=14, fontweight='bold', y=1.0)
plt.show()

print("=" * 60)
print("åŠ¨ä½œåºåˆ—å¯è§†åŒ–è¯´æ˜ï¼š")
print("=" * 60)
print("1. å·¦ä¸Šï¼šåŠ¨ä½œåºåˆ—æ—¶é—´è½´ï¼Œæ˜¾ç¤ºæ¯ä¸ªåŠ¨ä½œçš„æ—¶é—´ä½ç½®å’ŒæŒç»­æ—¶é—´")
print("2. å³ä¸Šï¼šåŠ¨ä½œæŒç»­æ—¶é—´ï¼Œæ˜¾ç¤ºæ¯ä¸ªåŠ¨ä½œçš„æŒç»­æ—¶é—´")
print("3. å·¦ä¸‹ï¼šåŠ¨ä½œé—´éš”æ—¶é—´ï¼Œæ˜¾ç¤ºåŠ¨ä½œä¹‹é—´çš„é—´éš”")
print("4. å³ä¸‹ï¼šåŠ¨ä½œåºåˆ—æµç¨‹å›¾ï¼Œæ˜¾ç¤ºåŠ¨ä½œçš„æ‰§è¡Œé¡ºåº")
print("=" * 60)
````

````python
# ============================================
# è‡ªå›å½’é¢„æµ‹å¯è§†åŒ–ï¼ˆç¤ºä¾‹ï¼šVLAä»»åŠ¡çš„è‡ªå›å½’åŠ¨ä½œé¢„æµ‹ï¼‰
# ============================================
np.random.seed(42)

# ç¤ºä¾‹ï¼šVLAä»»åŠ¡"æ‹¿èµ·æ¯å­æ”¾åˆ°æ¡Œå­ä¸Š"
task_name = "æ‹¿èµ·æ¯å­æ”¾åˆ°æ¡Œå­ä¸Š"
action_sequence = [
    'ç§»åŠ¨åˆ°æ¯å­ä½ç½®',
    'æŠ“å–æ¯å­',
    'ç§»åŠ¨åˆ°æ¡Œå­ä½ç½®',
    'æ”¾ä¸‹æ¯å­',
]
sequence_length = len(action_sequence)

# æ¨¡æ‹Ÿè‡ªå›å½’é¢„æµ‹è¿‡ç¨‹
# æ¯ä¸ªåŠ¨ä½œçš„é¢„æµ‹æ¦‚ç‡ï¼ˆç®€åŒ–ç‰ˆï¼‰
action_probs_history = []
for t in range(sequence_length):
    # æ¨¡æ‹Ÿé¢„æµ‹æ¦‚ç‡ï¼ˆæ ¹æ®å†å²åŠ¨ä½œè°ƒæ•´ï¼‰
    if t == 0:
        # ç¬¬ä¸€ä¸ªåŠ¨ä½œï¼šåŸºäºå¤šæ¨¡æ€ç‰¹å¾
        probs = np.array([0.05, 0.85, 0.05, 0.05])  # é«˜æ¦‚ç‡é¢„æµ‹"ç§»åŠ¨åˆ°æ¯å­ä½ç½®"
    elif t == 1:
        # ç¬¬äºŒä¸ªåŠ¨ä½œï¼šåŸºäºç¬¬ä¸€ä¸ªåŠ¨ä½œ
        probs = np.array([0.05, 0.10, 0.80, 0.05])  # é«˜æ¦‚ç‡é¢„æµ‹"æŠ“å–æ¯å­"
    elif t == 2:
        # ç¬¬ä¸‰ä¸ªåŠ¨ä½œï¼šåŸºäºå‰ä¸¤ä¸ªåŠ¨ä½œ
        probs = np.array([0.05, 0.05, 0.10, 0.80])  # é«˜æ¦‚ç‡é¢„æµ‹"ç§»åŠ¨åˆ°æ¡Œå­ä½ç½®"
    else:
        # ç¬¬å››ä¸ªåŠ¨ä½œï¼šåŸºäºå‰ä¸‰ä¸ªåŠ¨ä½œ
        probs = np.array([0.05, 0.05, 0.05, 0.85])  # é«˜æ¦‚ç‡é¢„æµ‹"æ”¾ä¸‹æ¯å­"
    
    # å½’ä¸€åŒ–
    probs = probs / probs.sum()
    action_probs_history.append(probs)

# å¯è§†åŒ–
fig, axes = plt.subplots(2, 2, figsize=(16, 12))

# å·¦ä¸Šï¼šè‡ªå›å½’é¢„æµ‹è¿‡ç¨‹ï¼ˆæ¯ä¸ªæ—¶é—´æ­¥çš„é¢„æµ‹æ¦‚ç‡ï¼‰
ax1 = axes[0, 0]
x_pos = np.arange(sequence_length)
width = 0.2
for t in range(sequence_length):
    predicted_idx = np.argmax(action_probs_history[t])
    colors = ['red' if i == predicted_idx else 'steelblue' for i in range(sequence_length)]
    bars = ax1.bar(x_pos + t * width, action_probs_history[t], width, 
                  color=colors, alpha=0.7, edgecolor='black', linewidth=1,
                  label=f'æ—¶é—´æ­¥{t+1}')
ax1.set_title('è‡ªå›å½’é¢„æµ‹è¿‡ç¨‹ï¼ˆæ¯ä¸ªæ—¶é—´æ­¥çš„é¢„æµ‹æ¦‚ç‡ï¼‰', fontsize=12, fontweight='bold')
ax1.set_xlabel('åŠ¨ä½œç´¢å¼•')
ax1.set_ylabel('é¢„æµ‹æ¦‚ç‡')
ax1.set_xticks(x_pos + width * 1.5)
ax1.set_xticklabels([f'åŠ¨ä½œ{i+1}' for i in range(sequence_length)])
ax1.set_ylim(0, 1.1)
ax1.legend(fontsize=9)
ax1.grid(True, alpha=0.3, axis='y')

# å³ä¸Šï¼šé¢„æµ‹åŠ¨ä½œåºåˆ—
ax2 = axes[0, 1]
y_pos = np.arange(sequence_length)
predicted_actions = [action_sequence[np.argmax(probs)] for probs in action_probs_history]
probs_max = [np.max(probs) for probs in action_probs_history]
colors_bar = ['red' if prob > 0.7 else 'orange' if prob > 0.5 else 'yellow' 
             for prob in probs_max]
bars = ax2.barh(y_pos, probs_max, color=colors_bar, alpha=0.7, 
               edgecolor='black', linewidth=2)
ax2.set_title('é¢„æµ‹åŠ¨ä½œåºåˆ—åŠå…¶ç½®ä¿¡åº¦', fontsize=12, fontweight='bold')
ax2.set_xlabel('é¢„æµ‹æ¦‚ç‡')
ax2.set_ylabel('åŠ¨ä½œæ­¥éª¤')
ax2.set_yticks(y_pos)
ax2.set_yticklabels([f'æ­¥éª¤{i+1}: {pred}' for i, pred in enumerate(predicted_actions)])
ax2.set_xlim(0, 1.1)
ax2.invert_yaxis()
ax2.grid(True, alpha=0.3, axis='x')

# æ·»åŠ æ•°å€¼æ ‡æ³¨
for i, (bar, prob) in enumerate(zip(bars, probs_max)):
    width_bar = bar.get_width()
    ax2.text(width_bar + 0.02, bar.get_y() + bar.get_height()/2,
             f'{prob:.2f}', ha='left', va='center', fontsize=10, fontweight='bold')

# å·¦ä¸‹ï¼šè‡ªå›å½’ä¾èµ–å…³ç³»
ax3 = axes[1, 0]
ax3.axis('off')
# ç»˜åˆ¶è‡ªå›å½’ä¾èµ–å…³ç³»å›¾
y_positions = [0.9, 0.7, 0.5, 0.3, 0.1]
labels = ['å¤šæ¨¡æ€ç‰¹å¾', 'åŠ¨ä½œ1', 'åŠ¨ä½œ2', 'åŠ¨ä½œ3', 'åŠ¨ä½œ4']
colors_dep = ['blue', 'green', 'orange', 'purple', 'red']

for i, (label, y, color) in enumerate(zip(labels, y_positions, colors_dep)):
    ax3.text(0.5, y, label, ha='center', va='center', fontsize=11, fontweight='bold',
            bbox=dict(boxstyle='round', facecolor=color, alpha=0.7))
    if i > 0:
        # ç»˜åˆ¶ä¾èµ–ç®­å¤´
        ax3.arrow(0.5, y_positions[i-1] - 0.05, 0, -0.08, head_width=0.03, 
                 head_length=0.02, fc='black', ec='black')
        # ç»˜åˆ¶ä»å¤šæ¨¡æ€ç‰¹å¾åˆ°æ¯ä¸ªåŠ¨ä½œçš„ç®­å¤´
        if i < len(labels) - 1:
            ax3.arrow(0.3, y_positions[0] - 0.05, 0.15, y - y_positions[0] + 0.05, 
                     head_width=0.02, head_length=0.02, fc='blue', ec='blue', 
                     alpha=0.5, linestyle='--')

ax3.set_xlim(0, 1)
ax3.set_ylim(0, 1)
ax3.set_title('è‡ªå›å½’ä¾èµ–å…³ç³»å›¾', fontsize=12, fontweight='bold')

# å³ä¸‹ï¼šé¢„æµ‹ç½®ä¿¡åº¦æ¼”åŒ–
ax4 = axes[1, 1]
time_steps = range(1, sequence_length + 1)
probs_max_list = [np.max(probs) for probs in action_probs_history]
ax4.plot(time_steps, probs_max_list, 'o-', linewidth=2, markersize=10, 
        color='red', alpha=0.7, label='é¢„æµ‹ç½®ä¿¡åº¦')
ax4.set_title('é¢„æµ‹ç½®ä¿¡åº¦æ¼”åŒ–', fontsize=12, fontweight='bold')
ax4.set_xlabel('æ—¶é—´æ­¥')
ax4.set_ylabel('é¢„æµ‹ç½®ä¿¡åº¦')
ax4.set_xticks(time_steps)
ax4.set_xticklabels([f'æ­¥éª¤{i}' for i in time_steps])
ax4.set_ylim(0, 1.1)
ax4.grid(True, alpha=0.3)
ax4.legend()

# æ·»åŠ æ•°å€¼æ ‡æ³¨
for t, prob in zip(time_steps, probs_max_list):
    ax4.text(t, prob + 0.03, f'{prob:.2f}', ha='center', va='bottom', 
            fontsize=10, fontweight='bold')

plt.tight_layout()
plt.suptitle(f'è‡ªå›å½’é¢„æµ‹å¯è§†åŒ–ï¼š{task_name}', fontsize=14, fontweight='bold', y=1.0)
plt.show()

print("=" * 60)
print("è‡ªå›å½’é¢„æµ‹å¯è§†åŒ–è¯´æ˜ï¼š")
print("=" * 60)
print("1. å·¦ä¸Šï¼šè‡ªå›å½’é¢„æµ‹è¿‡ç¨‹ï¼Œæ˜¾ç¤ºæ¯ä¸ªæ—¶é—´æ­¥çš„é¢„æµ‹æ¦‚ç‡")
print("2. å³ä¸Šï¼šé¢„æµ‹åŠ¨ä½œåºåˆ—åŠå…¶ç½®ä¿¡åº¦ï¼Œæ˜¾ç¤ºæ¯ä¸ªæ­¥éª¤çš„é¢„æµ‹åŠ¨ä½œå’Œç½®ä¿¡åº¦")
print("3. å·¦ä¸‹ï¼šè‡ªå›å½’ä¾èµ–å…³ç³»å›¾ï¼Œæ˜¾ç¤ºåŠ¨ä½œä¹‹é—´çš„ä¾èµ–å…³ç³»")
print("4. å³ä¸‹ï¼šé¢„æµ‹ç½®ä¿¡åº¦æ¼”åŒ–ï¼Œæ˜¾ç¤ºé¢„æµ‹ç½®ä¿¡åº¦éšæ—¶é—´æ­¥çš„å˜åŒ–")
print("=" * 60)
````
