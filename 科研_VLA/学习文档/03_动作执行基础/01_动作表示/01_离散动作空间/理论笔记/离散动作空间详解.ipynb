### 1. 离散动作空间 (Discrete Action Space)
- **中文名称**：离散动作空间
- **英文全称**：Discrete Action Space
- **定义**：离散动作空间是指将动作表示为离散类别的动作空间，是动作表示的一种方法。在离散动作空间中，每个动作都是一个类别，动作空间是有限的、可枚举的。离散动作空间的优势在于：1）简单直观：动作表示简单，易于理解和实现；2）分类预测：可以使用分类方法预测动作，训练简单；3）适合导航任务：适合需要选择离散动作的任务，如导航、游戏等。离散动作空间的劣势在于：1）表达能力有限：无法表示连续的动作参数；2）动作粒度固定：动作粒度由动作空间定义决定，无法灵活调整。离散动作空间在VLA中的应用包括导航任务、简单操作任务等，这些任务通常只需要选择离散的动作类型。离散动作空间是VLA动作表示的基础，通过离散动作空间，VLA模型能够将动作表示为离散类别，使用分类方法预测动作，然后执行相应的动作。离散动作空间的质量直接影响VLA模型的性能，好的离散动作空间设计能够帮助模型更好地预测动作，提高任务完成率。在VLA训练过程中，离散动作空间通常与分类预测方法一起使用，通过交叉熵损失函数训练模型，学习最适合VLA任务的动作表示方法。
- **核心组成**：离散动作空间的核心组成包括：1）动作类别定义：定义动作类别，如前进、后退、左转、右转、停止等；2）动作编码：将动作编码为类别索引或one-hot向量；3）动作空间大小：定义动作空间的大小，即动作类别的数量；4）动作映射：建立动作类别到实际动作的映射；5）动作选择：从动作空间中选择动作；6）动作执行：将选择的动作转换为实际机器人动作。离散动作空间通常使用分类方法预测动作，如使用softmax输出动作概率分布。离散动作空间的设计需要考虑任务特点，选择合适的动作类别和动作空间大小，以平衡表达能力和预测复杂度。
- **在VLA中的应用**：在VLA中，离散动作空间用于表示需要选择离散动作的任务。例如，导航任务可以使用离散动作空间，动作类别包括前进、后退、左转、右转、停止等。VLA模型使用分类方法预测动作类别，然后执行相应的动作。离散动作空间的优势在于简单直观，训练简单，适合简单的VLA任务。在VLA训练过程中，离散动作空间通常与分类预测方法一起使用，通过交叉熵损失函数训练模型。离散动作空间还可以用于动作序列预测，通过预测每个时间步的动作类别，生成动作序列。离散动作空间的结果可以作为VLA模型的输出，帮助模型选择动作类型，从而执行相应的动作。

- **直观理解**：想象离散动作空间就像选择菜单，菜单上有几个固定的选项，你只能选择其中一个。例如，导航任务的离散动作空间就像选择"前进"、"后退"、"左转"、"右转"、"停止"等选项。在VLA中，离散动作空间帮助模型选择动作类型，从而执行相应的动作。离散动作空间就像给动作贴上"类别标签"，帮助模型选择动作类型。

### 2. One-Hot编码 (One-Hot Encoding)
- **中文名称**：One-Hot编码
- **英文全称**：One-Hot Encoding
- **定义**：One-Hot编码是指将类别变量编码为二进制向量的编码方法，是离散动作空间中最常用的编码方法。在One-Hot编码中，每个类别对应一个向量，向量中只有一个位置为1，其他位置为0。例如，对于3个类别的动作空间，动作0编码为$[1, 0, 0]$，动作1编码为$[0, 1, 0]$，动作2编码为$[0, 0, 1]$。One-Hot编码的优势在于：1）简单直观：编码方法简单，易于理解和实现；2）类别独立：每个类别都有独立的表示，类别之间没有数值关系；3）适合分类：适合分类任务，可以直接用于softmax等分类函数。One-Hot编码的劣势在于：1）维度高：动作空间大时，向量维度很高；2）稀疏表示：向量中大部分位置为0，表示稀疏。One-Hot编码在VLA中的应用包括将离散动作编码为向量表示，这些表示将被用于动作预测模型。
- **核心组成**：One-Hot编码的核心组成包括：1）类别索引：将类别映射到索引，如动作0、动作1、动作2等；2）向量维度：定义向量的维度，等于动作空间的大小；3）编码规则：定义编码规则，即哪个位置为1；4）解码规则：定义解码规则，即如何从向量恢复类别；5）编码实现：实现编码函数，将类别转换为向量；6）解码实现：实现解码函数，将向量转换为类别。One-Hot编码通常使用简单的索引映射实现，对于类别$i$，编码为长度为$|A|$的向量，其中第$i$个位置为1，其他位置为0。
- **在VLA中的应用**：在VLA中，One-Hot编码用于将离散动作编码为向量表示。例如，对于导航任务的离散动作空间（前进、后退、左转、右转、停止），每个动作可以编码为one-hot向量，如前进编码为$[1, 0, 0, 0, 0]$，后退编码为$[0, 1, 0, 0, 0]$等。这些one-hot向量将被用于动作预测模型，模型输出动作概率分布，然后选择概率最高的动作。在VLA训练过程中，One-Hot编码通常与分类预测方法一起使用，通过交叉熵损失函数训练模型。One-Hot编码的结果可以作为VLA模型的输入或输出，帮助模型处理离散动作。
- **相关概念**：离散动作空间、类别索引、softmax、交叉熵损失
- **首次出现位置**：本文档第1.2节
- **深入学习**：参考本文档的One-Hot编码详细讲解部分
- **直观理解**：想象One-Hot编码就像给每个动作贴上一个"标签"，这个标签是一个向量，向量中只有一个位置是"亮"的（值为1），其他位置都是"暗"的（值为0）。例如，对于5个动作，前进的标签是$[1, 0, 0, 0, 0]$，后退的标签是$[0, 1, 0, 0, 0]$等。在VLA中，One-Hot编码帮助模型识别动作类型，从而执行相应的动作。One-Hot编码就像给动作贴上"身份标签"，帮助模型识别动作类型。

## 1. 离散动作空间的基本原理

### 1.1 什么是离散动作空间

**直观理解**：想象离散动作空间就像选择菜单，菜单上有几个固定的选项，你只能选择其中一个。

**定义**：离散动作空间是指将动作表示为离散类别的动作空间，是动作表示的一种方法。

### 1.2 离散动作空间的数学表示

离散动作空间的数学表示为：

$$\mathcal{A} = \{a_1, a_2, \ldots, a_{|\mathcal{A}|}\}$$

其中：
- $\mathcal{A}$ 是动作空间
- $a_i$ 是第$i$个动作类别
- $|\mathcal{A}|$ 是动作空间的大小（动作类别的数量）

每个动作 $a$ 可以表示为类别索引：

$$a \in \{0, 1, 2, \ldots, |\mathcal{A}|-1\}$$

或者使用one-hot编码：

$$a = [0, 0, \ldots, 1, \ldots, 0] \in \{0, 1\}^{|\mathcal{A}|}$$

其中只有第 $a$ 个位置为1。

### 1.3 离散动作空间的可视化

## 4. 离散动作空间的预测方法详解

### 4.1 动作预测的数学原理

在VLA中，离散动作空间的预测本质上是一个分类问题：给定多模态特征，预测应该执行哪个动作类别。

#### 4.1.1 从多模态特征到动作概率

**步骤1：理解多模态特征**

多模态特征是指融合了视觉特征和语言特征的特征表示：

$$f_{multimodal} \in \mathbb{R}^{d}$$

其中$d$是特征维度。

**步骤2：理解动作预测网络**

动作预测网络是一个分类网络，将多模态特征映射到动作概率分布：

$$P(a|f_{multimodal}) = \text{softmax}(W \cdot f_{multimodal} + b)$$

其中：
- $W \in \mathbb{R}^{|\mathcal{A}| \times d}$是权重矩阵
- $b \in \mathbb{R}^{|\mathcal{A}|}$是偏置向量
- $P(a|f_{multimodal}) \in \mathbb{R}^{|\mathcal{A}|}$是动作概率分布

**步骤3：理解Softmax函数**

Softmax函数将logits转换为概率分布：

$$\text{softmax}(z)_i = \frac{\exp(z_i)}{\sum_{j=1}^{|\mathcal{A}|} \exp(z_j)}$$

其中$z = W \cdot f_{multimodal} + b$是logits。

#### 4.1.2 Softmax的数学推导详解

**步骤1：理解指数函数**

指数函数$\exp(x) = e^x$的性质：
- 单调递增：$x_1 < x_2 \Rightarrow \exp(x_1) < \exp(x_2)$
- 非负性：$\exp(x) > 0$对所有$x$成立
- 放大差异：大的值会被放大，小的值会被缩小

**步骤2：理解归一化**

归一化是指将一组值转换为概率分布，使得所有值的和为1：

$$\sum_{i=1}^{|\mathcal{A}|} P(a_i|f_{multimodal}) = 1$$

**步骤3：理解Softmax的推导**

Softmax的推导过程：

1. **计算logits**：$z = W \cdot f_{multimodal} + b$
2. **应用指数函数**：$\exp(z_i)$对所有$i$
3. **归一化**：$\frac{\exp(z_i)}{\sum_{j=1}^{|\mathcal{A}|} \exp(z_j)}$

**为什么使用Softmax**：
- **概率分布**：输出是概率分布，所有值的和为1
- **非负性**：所有值都是非负的
- **可微性**：Softmax是可微的，便于梯度下降
- **放大差异**：指数函数会放大logits之间的差异

#### 4.1.3 具体计算示例

**示例：预测动作概率分布**

假设：
- 动作空间大小：$|\mathcal{A}| = 5$
- 多模态特征维度：$d = 128$
- 权重矩阵：$W \in \mathbb{R}^{5 \times 128}$（随机初始化）
- 偏置向量：$b \in \mathbb{R}^{5}$（随机初始化）
- 多模态特征：$f_{multimodal} \in \mathbb{R}^{128}$（随机生成）

**步骤1：计算logits**

$$z = W \cdot f_{multimodal} + b$$

假设计算结果为：
$$z = [2.5, 0.3, -0.8, 0.1, 1.2]$$

**步骤2：应用指数函数**

$$\exp(z) = [\exp(2.5), \exp(0.3), \exp(-0.8), \exp(0.1), \exp(1.2)]$$

计算：
- $\exp(2.5) \approx 12.182$
- $\exp(0.3) \approx 1.350$
- $\exp(-0.8) \approx 0.449$
- $\exp(0.1) \approx 1.105$
- $\exp(1.2) \approx 3.320$

所以：
$$\exp(z) \approx [12.182, 1.350, 0.449, 1.105, 3.320]$$

**步骤3：归一化**

归一化因子：
$$Z = \sum_{i=1}^{5} \exp(z_i) \approx 12.182 + 1.350 + 0.449 + 1.105 + 3.320 = 18.406$$

概率分布：
$$P(a|f_{multimodal}) = \frac{\exp(z)}{Z} \approx [0.662, 0.073, 0.024, 0.060, 0.180]$$

**步骤4：选择动作**

选择概率最高的动作：
$$\hat{a} = \arg\max_i P(a_i|f_{multimodal}) = 0$$

即预测动作为"前进"（索引0）。

### 4.2 交叉熵损失函数详解

#### 4.2.1 什么是交叉熵损失

交叉熵损失是分类任务中最常用的损失函数，用于衡量预测概率分布和真实标签之间的差异。

**数学定义**：

对于真实标签$y$（one-hot编码）和预测概率分布$P(a|f_{multimodal})$，交叉熵损失定义为：

$$\mathcal{L}_{CE} = -\sum_{i=1}^{|\mathcal{A}|} y_i \log(P(a_i|f_{multimodal}))$$

其中$y_i$是真实标签的one-hot编码。

#### 4.2.2 交叉熵损失的数学推导

**步骤1：理解信息熵**

信息熵是衡量信息量的指标：

$$H(X) = -\sum_{i=1}^{n} P(x_i) \log(P(x_i))$$

**步骤2：理解交叉熵**

交叉熵是衡量两个概率分布之间差异的指标：

$$H(P, Q) = -\sum_{i=1}^{n} P(x_i) \log(Q(x_i))$$

其中$P$是真实分布，$Q$是预测分布。

**步骤3：理解分类任务中的交叉熵**

在分类任务中：
- 真实分布$P$：one-hot编码（只有一个位置为1）
- 预测分布$Q$：模型输出的概率分布

所以交叉熵损失为：

$$\mathcal{L}_{CE} = -\sum_{i=1}^{|\mathcal{A}|} y_i \log(P(a_i|f_{multimodal}))$$

由于$y$是one-hot编码，只有一个位置为1，所以：

$$\mathcal{L}_{CE} = -\log(P(a_{true}|f_{multimodal}))$$

其中$a_{true}$是真实动作类别。

#### 4.2.3 具体计算示例

**示例：计算交叉熵损失**

假设：
- 真实动作：左转（索引2，one-hot编码：$[0, 0, 1, 0, 0]$）
- 预测概率分布：$P(a|f_{multimodal}) = [0.662, 0.073, 0.024, 0.060, 0.180]$

**步骤1：计算交叉熵损失**

$$\mathcal{L}_{CE} = -\sum_{i=1}^{5} y_i \log(P(a_i|f_{multimodal}))$$

由于$y$是one-hot编码，只有$y_2 = 1$，其他为0，所以：

$$\mathcal{L}_{CE} = -\log(P(a_2|f_{multimodal})) = -\log(0.024) \approx 3.729$$

**步骤2：理解损失的含义**

损失值越大，说明预测概率越小，模型预测越不准确。在这个例子中，真实动作是"左转"，但模型预测"左转"的概率只有0.024，所以损失很大。

下面我们通过代码可视化动作预测的过程：

离散动作空间在VLA中的重要性体现在：

````python
# ============================================
# 动作预测可视化（示例：VLA导航任务的动作预测）
# ============================================
np.random.seed(42)

# 示例：VLA导航任务的离散动作空间
action_space = {
    0: '前进',
    1: '后退',
    2: '左转',
    3: '右转',
    4: '停止',
}
action_space_size = len(action_space)

# 模拟多模态特征
d_model = 128
multimodal_features = np.random.randn(d_model)

# 模拟权重矩阵和偏置向量
W = np.random.randn(action_space_size, d_model) * 0.1
b = np.random.randn(action_space_size) * 0.1

# 计算logits
logits = W @ multimodal_features + b

# 计算softmax概率
exp_logits = np.exp(logits)
probs = exp_logits / np.sum(exp_logits)

# 预测动作
predicted_action = np.argmax(probs)
true_action = 2  # 假设真实动作是"左转"

# 计算交叉熵损失
true_one_hot = np.zeros(action_space_size)
true_one_hot[true_action] = 1
ce_loss = -np.sum(true_one_hot * np.log(probs + 1e-10))  # 加小值避免log(0)

# 可视化
fig, axes = plt.subplots(2, 2, figsize=(16, 12))

# 左上：Logits分布
ax1 = axes[0, 0]
action_names = list(action_space.values())
colors = ['red' if i == predicted_action else 'steelblue' for i in range(action_space_size)]
bars = ax1.bar(range(action_space_size), logits, color=colors, 
               edgecolor='black', linewidth=2, alpha=0.7)
ax1.set_title('Logits分布（Softmax之前）', fontsize=12, fontweight='bold')
ax1.set_xlabel('动作类别')
ax1.set_ylabel('Logits值')
ax1.set_xticks(range(action_space_size))
ax1.set_xticklabels(action_names, rotation=45, ha='right')
ax1.grid(True, alpha=0.3, axis='y')
ax1.axhline(y=0, color='black', linestyle='--', linewidth=1)

# 添加数值标注
for i, (bar, logit) in enumerate(zip(bars, logits)):
    height = bar.get_height()
    ax1.text(bar.get_x() + bar.get_width()/2., height + 0.1 if height >= 0 else height - 0.15,
             f'{logit:.2f}', ha='center', va='bottom' if height >= 0 else 'top', 
             fontsize=10, fontweight='bold')

# 右上：Softmax概率分布
ax2 = axes[0, 1]
colors_prob = ['red' if i == predicted_action else 'steelblue' for i in range(action_space_size)]
bars = ax2.bar(range(action_space_size), probs, color=colors_prob, 
               edgecolor='black', linewidth=2, alpha=0.7)
ax2.set_title(f'Softmax概率分布（预测动作：{action_names[predicted_action]}）', 
             fontsize=12, fontweight='bold')
ax2.set_xlabel('动作类别')
ax2.set_ylabel('概率')
ax2.set_xticks(range(action_space_size))
ax2.set_xticklabels(action_names, rotation=45, ha='right')
ax2.set_ylim(0, max(probs) * 1.2)
ax2.grid(True, alpha=0.3, axis='y')

# 添加数值标注
for i, (bar, prob) in enumerate(zip(bars, probs)):
    height = bar.get_height()
    ax2.text(bar.get_x() + bar.get_width()/2., height + 0.01,
             f'{prob:.3f}', ha='center', va='bottom', fontsize=10, fontweight='bold')

# 左下：真实标签 vs 预测概率
ax3 = axes[1, 0]
x_pos = np.arange(action_space_size)
width = 0.35
bars1 = ax3.bar(x_pos - width/2, true_one_hot, width, label='真实标签', 
               color='green', alpha=0.7, edgecolor='black', linewidth=1)
bars2 = ax3.bar(x_pos + width/2, probs, width, label='预测概率', 
               color='orange', alpha=0.7, edgecolor='black', linewidth=1)
ax3.set_title('真实标签 vs 预测概率', fontsize=12, fontweight='bold')
ax3.set_xlabel('动作类别')
ax3.set_ylabel('值')
ax3.set_xticks(x_pos)
ax3.set_xticklabels(action_names, rotation=45, ha='right')
ax3.set_ylim(0, 1.1)
ax3.legend()
ax3.grid(True, alpha=0.3, axis='y')

# 右下：交叉熵损失
ax4 = axes[1, 1]
ax4.axis('off')
ax4.text(0.5, 0.8, '交叉熵损失计算', fontsize=14, fontweight='bold', 
        ha='center', va='center', bbox=dict(boxstyle='round', facecolor='lightblue', alpha=0.5))
y_pos = 0.6
ax4.text(0.5, y_pos, f'真实动作：{action_names[true_action]}', 
        fontsize=12, ha='center', va='center',
        bbox=dict(boxstyle='round', facecolor='lightgreen', alpha=0.5))
y_pos -= 0.15
ax4.text(0.5, y_pos, f'预测动作：{action_names[predicted_action]}', 
        fontsize=12, ha='center', va='center',
        bbox=dict(boxstyle='round', facecolor='lightyellow', alpha=0.5))
y_pos -= 0.15
ax4.text(0.5, y_pos, f'交叉熵损失：{ce_loss:.4f}', 
        fontsize=12, ha='center', va='center',
        bbox=dict(boxstyle='round', facecolor='lightcoral', alpha=0.5))
y_pos -= 0.15
ax4.text(0.5, y_pos, f'预测概率：{probs[true_action]:.4f}', 
        fontsize=12, ha='center', va='center',
        bbox=dict(boxstyle='round', facecolor='lightgray', alpha=0.5))

plt.tight_layout()
plt.suptitle('动作预测可视化：VLA导航任务的动作预测', fontsize=14, fontweight='bold', y=1.0)
plt.show()

print("=" * 60)
print("动作预测可视化说明：")
print("=" * 60)
print("1. 左上：Logits分布，显示Softmax之前的原始分数")
print("2. 右上：Softmax概率分布，显示每个动作的预测概率")
print("3. 左下：真实标签 vs 预测概率，对比真实动作和预测概率")
print("4. 右下：交叉熵损失，显示损失值和相关信息")
print("=" * 60)
````

## 5. 动作空间设计详解

### 5.1 动作类别设计

#### 5.1.1 动作类别的设计原则

动作类别的设计需要考虑以下原则：

1. **任务相关性**：
   - 动作类别必须与任务相关
   - 每个动作类别都应该对任务完成有贡献
   - 避免冗余的动作类别

2. **互斥性**：
   - 动作类别之间应该是互斥的
   - 每个时刻只能执行一个动作
   - 避免动作类别之间的重叠

3. **完整性**：
   - 动作类别应该覆盖任务所需的所有动作
   - 不能遗漏关键动作
   - 确保任务可以完成

4. **粒度平衡**：
   - 动作粒度不能太粗（无法完成任务）
   - 动作粒度不能太细（增加预测复杂度）
   - 需要平衡表达能力和预测复杂度

#### 5.1.2 动作类别设计示例

**示例1：导航任务**

对于导航任务，动作类别可以设计为：

```python
navigation_actions = {
    0: '前进',
    1: '后退',
    2: '左转',
    3: '右转',
    4: '停止',
}
```

**设计理由**：
- **任务相关性**：导航任务需要移动和转向
- **互斥性**：每个动作都是独立的
- **完整性**：覆盖了导航所需的基本动作
- **粒度平衡**：粒度适中，不会太粗或太细

**示例2：抓取任务**

对于抓取任务，动作类别可以设计为：

```python
grasping_actions = {
    0: '移动到目标',
    1: '抓取',
    2: '释放',
    3: '抬起',
    4: '放下',
}
```

**设计理由**：
- **任务相关性**：抓取任务需要移动、抓取、释放等动作
- **互斥性**：每个动作都是独立的
- **完整性**：覆盖了抓取任务所需的基本动作
- **粒度平衡**：粒度适中，适合简单抓取任务

### 5.2 动作空间大小设计

#### 5.2.1 动作空间大小的影响

动作空间大小对模型性能的影响：

1. **表达能力**：
   - 动作空间越大，表达能力越强
   - 可以表示更多的动作类型
   - 但预测复杂度也越高

2. **预测复杂度**：
   - 动作空间越大，预测复杂度越高
   - 需要更多的训练数据
   - 训练时间更长

3. **泛化能力**：
   - 动作空间太大可能导致过拟合
   - 动作空间太小可能导致欠拟合
   - 需要平衡表达能力和泛化能力

#### 5.2.2 动作空间大小设计原则

动作空间大小的设计原则：

1. **最小化原则**：
   - 使用最小的动作空间完成任务
   - 避免不必要的动作类别
   - 减少预测复杂度

2. **任务需求**：
   - 根据任务需求确定动作空间大小
   - 简单任务使用小的动作空间
   - 复杂任务可能需要更大的动作空间

3. **实验验证**：
   - 通过实验验证动作空间大小
   - 比较不同大小的动作空间的性能
   - 选择最优的动作空间大小

#### 5.2.3 动作空间大小设计示例

**示例：不同任务的动作空间大小**

| 任务类型 | 动作空间大小 | 动作类别示例 |
|---------|------------|------------|
| 简单导航 | 5 | 前进、后退、左转、右转、停止 |
| 复杂导航 | 9 | 前进、后退、左转、右转、停止、左前、右前、左后、右后 |
| 简单抓取 | 5 | 移动到目标、抓取、释放、抬起、放下 |
| 复杂操作 | 10+ | 多种组合动作 |

````python
# ============================================
# 动作空间设计可视化（示例：不同动作空间大小的影响）
# ============================================
np.random.seed(42)

# 不同任务的动作空间设计
task_designs = {
    '简单导航': {
        'size': 5,
        'actions': ['前进', '后退', '左转', '右转', '停止'],
        'complexity': 5 * np.log(5),
    },
    '复杂导航': {
        'size': 9,
        'actions': ['前进', '后退', '左转', '右转', '停止', '左前', '右前', '左后', '右后'],
        'complexity': 9 * np.log(9),
    },
    '简单抓取': {
        'size': 5,
        'actions': ['移动到目标', '抓取', '释放', '抬起', '放下'],
        'complexity': 5 * np.log(5),
    },
    '复杂操作': {
        'size': 12,
        'actions': ['移动到目标', '抓取', '释放', '抬起', '放下', '旋转', '推', '拉', '打开', '关闭', '组合1', '组合2'],
        'complexity': 12 * np.log(12),
    },
}

# 可视化
fig, axes = plt.subplots(2, 2, figsize=(16, 12))

# 左上：动作空间大小对比
ax1 = axes[0, 0]
task_names = list(task_designs.keys())
sizes = [task_designs[t]['size'] for t in task_names]
colors = ['red', 'blue', 'green', 'orange']
bars = ax1.bar(range(len(task_names)), sizes, color=colors, 
               edgecolor='black', linewidth=2, alpha=0.7)
ax1.set_title('不同任务的动作空间大小', fontsize=12, fontweight='bold')
ax1.set_xlabel('任务类型')
ax1.set_ylabel('动作空间大小')
ax1.set_xticks(range(len(task_names)))
ax1.set_xticklabels(task_names, rotation=45, ha='right')
ax1.set_ylim(0, max(sizes) * 1.2)
ax1.grid(True, alpha=0.3, axis='y')

# 添加数值标注
for i, (bar, size) in enumerate(zip(bars, sizes)):
    height = bar.get_height()
    ax1.text(bar.get_x() + bar.get_width()/2., height + 0.3,
             f'{size}', ha='center', va='bottom', fontsize=10, fontweight='bold')

# 右上：预测复杂度对比
ax2 = axes[0, 1]
complexities = [task_designs[t]['complexity'] for t in task_names]
bars = ax2.bar(range(len(task_names)), complexities, color=colors, 
               edgecolor='black', linewidth=2, alpha=0.7)
ax2.set_title('不同任务的预测复杂度', fontsize=12, fontweight='bold')
ax2.set_xlabel('任务类型')
ax2.set_ylabel('预测复杂度（相对值）')
ax2.set_xticks(range(len(task_names)))
ax2.set_xticklabels(task_names, rotation=45, ha='right')
ax2.set_ylim(0, max(complexities) * 1.2)
ax2.grid(True, alpha=0.3, axis='y')

# 添加数值标注
for i, (bar, comp) in enumerate(zip(bars, complexities)):
    height = bar.get_height()
    ax2.text(bar.get_x() + bar.get_width()/2., height + 1,
             f'{comp:.1f}', ha='center', va='bottom', fontsize=10, fontweight='bold')

# 左下：动作空间大小 vs 预测复杂度
ax3 = axes[1, 0]
space_sizes = np.arange(3, 21)
complexities_curve = space_sizes * np.log(space_sizes)
ax3.plot(space_sizes, complexities_curve, 'o-', linewidth=2, markersize=6, 
        color='red', alpha=0.7, label='预测复杂度')
ax3.set_title('动作空间大小 vs 预测复杂度', fontsize=12, fontweight='bold')
ax3.set_xlabel('动作空间大小')
ax3.set_ylabel('预测复杂度（相对值）')
ax3.grid(True, alpha=0.3)
ax3.legend()
ax3.set_xlim(2, 21)

# 标注不同任务的位置
for task_name, task_info in task_designs.items():
    ax3.scatter(task_info['size'], task_info['complexity'], 
               s=200, color=colors[task_names.index(task_name)], 
               edgecolor='black', linewidth=2, zorder=5)
    ax3.annotate(task_name, (task_info['size'], task_info['complexity']),
                xytext=(5, 5), textcoords='offset points', fontsize=9, fontweight='bold')

# 右下：动作类别设计示例
ax4 = axes[1, 1]
ax4.axis('off')
y_pos = 0.95
for task_name, task_info in task_designs.items():
    y_pos -= 0.2
    ax4.text(0.05, y_pos, f'{task_name} ({task_info["size"]}个动作):', 
            fontsize=11, fontweight='bold', ha='left', va='top')
    y_pos -= 0.1
    actions_str = ', '.join(task_info['actions'][:6])  # 只显示前6个
    if len(task_info['actions']) > 6:
        actions_str += '...'
    ax4.text(0.1, y_pos, actions_str, fontsize=10, ha='left', va='top',
            bbox=dict(boxstyle='round', facecolor='lightblue', alpha=0.3))
    y_pos -= 0.05

ax4.set_xlim(0, 1)
ax4.set_ylim(0, 1)
ax4.set_title('动作类别设计示例', fontsize=12, fontweight='bold')

plt.tight_layout()
plt.suptitle('动作空间设计可视化：不同任务的动作空间设计', fontsize=14, fontweight='bold', y=1.0)
plt.show()

print("=" * 60)
print("动作空间设计可视化说明：")
print("=" * 60)
print("1. 左上：不同任务的动作空间大小对比")
print("2. 右上：不同任务的预测复杂度对比")
print("3. 左下：动作空间大小 vs 预测复杂度的关系曲线")
print("4. 右下：不同任务的动作类别设计示例")
print("=" * 60)
````

def one_hot_encode(action_idx, space_size):
