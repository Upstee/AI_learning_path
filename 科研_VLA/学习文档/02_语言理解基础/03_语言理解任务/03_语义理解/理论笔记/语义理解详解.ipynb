```python
# ============================================
# æ„å›¾è¯†åˆ«å¯è§†åŒ–ï¼ˆç¤ºä¾‹ï¼šVLAæŒ‡ä»¤çš„æ„å›¾è¯†åˆ«ï¼‰
# ============================================
np.random.seed(42)

# ç¤ºä¾‹ï¼šVLAæŒ‡ä»¤
instructions = [
    "æ‹¿èµ·æ¡Œå­ä¸Šçš„çº¢è‰²æ¯å­",
    "æŠŠæ¯å­æ”¾åœ¨æ¶å­ä¸Š",
    "ç§»åŠ¨åˆ°å¨æˆ¿",
    "æ‰“å¼€æŠ½å±‰",
]

# æ„å›¾ç±»åˆ«
intent_types = ['æŠ“å–', 'æ”¾ç½®', 'å¯¼èˆª', 'æ“ä½œ']
num_intents = len(intent_types)

# æ¨¡æ‹Ÿæ„å›¾è¯†åˆ«çš„æ¦‚ç‡ï¼ˆç®€åŒ–ç‰ˆï¼‰
np.random.seed(42)
intent_probs = np.random.rand(len(instructions), num_intents)
intent_probs = intent_probs / intent_probs.sum(axis=1, keepdims=True)

# è®¾ç½®åˆç†çš„æ¦‚ç‡ï¼ˆæ¨¡æ‹Ÿæ¨¡å‹é¢„æµ‹ï¼‰
intent_probs[0, 0] = 0.85  # "æ‹¿èµ·" -> æŠ“å–
intent_probs[0, 1:] = [0.05, 0.05, 0.05]
intent_probs[1, 1] = 0.80  # "æ”¾åœ¨" -> æ”¾ç½®
intent_probs[1, [0, 2, 3]] = [0.05, 0.10, 0.05]
intent_probs[2, 2] = 0.90  # "ç§»åŠ¨åˆ°" -> å¯¼èˆª
intent_probs[2, [0, 1, 3]] = [0.03, 0.03, 0.04]
intent_probs[3, 3] = 0.75  # "æ‰“å¼€" -> æ“ä½œ
intent_probs[3, [0, 1, 2]] = [0.08, 0.10, 0.07]

# é‡æ–°å½’ä¸€åŒ–
intent_probs = intent_probs / intent_probs.sum(axis=1, keepdims=True)

# å¯è§†åŒ–
fig, axes = plt.subplots(2, 2, figsize=(16, 12))
axes = axes.flatten()

for idx, (instruction, probs) in enumerate(zip(instructions, intent_probs)):
    ax = axes[idx]
    predicted_intent = np.argmax(probs)
    colors = ['red' if i == predicted_intent else 'steelblue' for i in range(num_intents)]
    bars = ax.bar(range(num_intents), probs, color=colors, 
                 edgecolor='black', linewidth=2, alpha=0.7)
    ax.set_title(f'æŒ‡ä»¤: {instruction}\né¢„æµ‹æ„å›¾: {intent_types[predicted_intent]}', 
                fontsize=11, fontweight='bold')
    ax.set_xlabel('æ„å›¾ç±»åˆ«')
    ax.set_ylabel('æ¦‚ç‡')
    ax.set_xticks(range(num_intents))
    ax.set_xticklabels(intent_types, rotation=45, ha='right')
    ax.set_ylim(0, 1.1)
    ax.grid(True, alpha=0.3, axis='y')
    
    # æ ‡æ³¨æ¦‚ç‡å€¼
    for i, (bar, prob) in enumerate(zip(bars, probs)):
        height = bar.get_height()
        ax.text(bar.get_x() + bar.get_width()/2., height + 0.02,
               f'{prob:.2f}', ha='center', va='bottom', fontsize=10, fontweight='bold')

plt.tight_layout()
plt.suptitle('æ„å›¾è¯†åˆ«å¯è§†åŒ–ï¼šVLAæŒ‡ä»¤çš„æ„å›¾è¯†åˆ«', fontsize=14, fontweight='bold', y=1.0)
plt.show()

print("=" * 60)
print("æ„å›¾è¯†åˆ«å¯è§†åŒ–è¯´æ˜ï¼š")
print("=" * 60)
print("1. æ¯ä¸ªå­å›¾æ˜¾ç¤ºä¸€ä¸ªæŒ‡ä»¤çš„æ„å›¾è¯†åˆ«ç»“æœ")
print("2. çº¢è‰²æŸ±è¡¨ç¤ºé¢„æµ‹æ„å›¾ï¼Œè“è‰²æŸ±è¡¨ç¤ºå…¶ä»–æ„å›¾")
print("3. æ„å›¾è¯†åˆ«çš„ç›®æ ‡æ˜¯è¯†åˆ«æŒ‡ä»¤è¦å®Œæˆä»€ä¹ˆä»»åŠ¡")
print("=" * 60)
```

```python
# ============================================
# è¯­ä¹‰ç›¸ä¼¼åº¦å¯è§†åŒ–ï¼ˆç¤ºä¾‹ï¼šVLAæŒ‡ä»¤çš„è¯­ä¹‰ç›¸ä¼¼åº¦ï¼‰
# ============================================
np.random.seed(42)

# ç¤ºä¾‹ï¼šVLAæŒ‡ä»¤å¯¹
instruction_pairs = [
    ("æ‹¿èµ·æ¯å­", "æŠ“å–æ¯å­", 0.92),  # é«˜ç›¸ä¼¼åº¦
    ("æ‹¿èµ·æ¯å­", "æ”¾ä¸‹æ¯å­", 0.35),  # ä½ç›¸ä¼¼åº¦
    ("ç§»åŠ¨åˆ°å¨æˆ¿", "å‰å¾€å¨æˆ¿", 0.88),  # é«˜ç›¸ä¼¼åº¦
    ("æ‰“å¼€æŠ½å±‰", "å…³é—­æŠ½å±‰", 0.42),  # ä½ç›¸ä¼¼åº¦
    ("æ‹¿èµ·çº¢è‰²æ¯å­", "æŠ“å–çº¢è‰²æ¯å­", 0.95),  # å¾ˆé«˜ç›¸ä¼¼åº¦
]

# æ¨¡æ‹Ÿè¯­ä¹‰å‘é‡ï¼ˆç®€åŒ–ç‰ˆï¼Œä½¿ç”¨ä½ç»´å‘é‡ä¾¿äºå¯è§†åŒ–ï¼‰
d_model = 768
d_vis = 2  # ç”¨äºå¯è§†åŒ–çš„ç»´åº¦ï¼ˆPCAåˆ°2Dï¼‰

# ç”Ÿæˆè¯­ä¹‰å‘é‡ï¼ˆç¡®ä¿ç›¸ä¼¼åº¦ç¬¦åˆé¢„æœŸï¼‰
np.random.seed(42)
vectors = {}
for text1, text2, target_sim in instruction_pairs:
    # ç”ŸæˆåŸºç¡€å‘é‡
    base_vec = np.random.randn(d_model)
    base_vec = base_vec / np.linalg.norm(base_vec)
    
    # æ–‡æœ¬1çš„å‘é‡
    vec1 = base_vec + np.random.randn(d_model) * 0.1
    vec1 = vec1 / np.linalg.norm(vec1)
    vectors[text1] = vec1
    
    # æ–‡æœ¬2çš„å‘é‡ï¼ˆä¸æ–‡æœ¬1ç›¸ä¼¼ï¼‰
    # é€šè¿‡è°ƒæ•´ç›¸ä¼¼åº¦æ¥æ§åˆ¶å‘é‡ä¹‹é—´çš„è§’åº¦
    angle = np.arccos(np.clip(target_sim, -1, 1))
    # ç”Ÿæˆä¸vec1æˆangleè§’åº¦çš„å‘é‡
    vec2 = base_vec + np.random.randn(d_model) * (0.1 + (1 - target_sim) * 0.5)
    vec2 = vec2 / np.linalg.norm(vec2)
    # è°ƒæ•´è§’åº¦ä½¿å…¶ç¬¦åˆç›®æ ‡ç›¸ä¼¼åº¦
    if target_sim < 0.5:
        vec2 = -vec2 * 0.3 + vec1 * 0.7  # é™ä½ç›¸ä¼¼åº¦
    vectors[text2] = vec2

# è®¡ç®—å®é™…ç›¸ä¼¼åº¦
actual_sims = []
for text1, text2, target_sim in instruction_pairs:
    vec1 = vectors[text1]
    vec2 = vectors[text2]
    actual_sim = np.dot(vec1, vec2)
    actual_sims.append(actual_sim)

# å¯è§†åŒ–
fig, axes = plt.subplots(2, 2, figsize=(16, 12))

# å·¦ä¸Šï¼šè¯­ä¹‰ç›¸ä¼¼åº¦å¯¹æ¯”
ax1 = axes[0, 0]
pair_labels = [f'{t1}\\n{t2}' for t1, t2, _ in instruction_pairs]
x_pos = np.arange(len(instruction_pairs))
width = 0.35

bars1 = ax1.bar(x_pos - width/2, [s for _, _, s in instruction_pairs], width, 
               label='ç›®æ ‡ç›¸ä¼¼åº¦', color='steelblue', alpha=0.7, 
               edgecolor='black', linewidth=1.5)
bars2 = ax1.bar(x_pos + width/2, actual_sims, width, 
               label='å®é™…ç›¸ä¼¼åº¦', color='lightcoral', alpha=0.7, 
               edgecolor='black', linewidth=1.5)

ax1.set_title('è¯­ä¹‰ç›¸ä¼¼åº¦å¯¹æ¯”', fontsize=12, fontweight='bold')
ax1.set_xlabel('æŒ‡ä»¤å¯¹')
ax1.set_ylabel('ç›¸ä¼¼åº¦')
ax1.set_xticks(x_pos)
ax1.set_xticklabels(pair_labels, rotation=45, ha='right', fontsize=9)
ax1.set_ylim([0, 1.1])
ax1.grid(True, alpha=0.3, axis='y')
ax1.legend()

# æ·»åŠ æ•°å€¼æ ‡æ³¨
for i, (bar1, bar2, target, actual) in enumerate(zip(bars1, bars2, 
    [s for _, _, s in instruction_pairs], actual_sims)):
    ax1.text(bar1.get_x() + bar1.get_width()/2., target + 0.02,
             f'{target:.2f}', ha='center', va='bottom', fontsize=9, fontweight='bold')
    ax1.text(bar2.get_x() + bar2.get_width()/2., actual + 0.02,
             f'{actual:.2f}', ha='center', va='bottom', fontsize=9, fontweight='bold')

# å³ä¸Šï¼šè¯­ä¹‰å‘é‡ç©ºé—´å¯è§†åŒ–ï¼ˆPCAåˆ°2Dï¼‰
ax2 = axes[0, 1]
all_texts = []
all_vectors = []
for text1, text2, _ in instruction_pairs:
    all_texts.extend([text1, text2])
    all_vectors.extend([vectors[text1], vectors[text2]])

all_vectors = np.array(all_vectors)
pca = PCA(n_components=2)
vectors_2d = pca.fit_transform(all_vectors)

# ç»˜åˆ¶å‘é‡
colors = ['red', 'blue', 'green', 'orange', 'purple'] * 2
for i, (text, vec_2d, color) in enumerate(zip(all_texts, vectors_2d, colors)):
    ax2.scatter(vec_2d[0], vec_2d[1], s=200, marker='o', color=color, 
               edgecolors='black', linewidths=2, alpha=0.7)
    ax2.text(vec_2d[0] + 0.05, vec_2d[1] + 0.05, text, fontsize=9, fontweight='bold')

# è¿æ¥ç›¸ä¼¼åº¦é«˜çš„æŒ‡ä»¤å¯¹
for i, (text1, text2, sim) in enumerate(instruction_pairs):
    idx1 = all_texts.index(text1)
    idx2 = all_texts.index(text2)
    if sim > 0.7:
        ax2.plot([vectors_2d[idx1, 0], vectors_2d[idx2, 0]], 
                [vectors_2d[idx1, 1], vectors_2d[idx2, 1]], 
                'g--', alpha=0.5, linewidth=2, label='é«˜ç›¸ä¼¼åº¦' if i == 0 else '')
    else:
        ax2.plot([vectors_2d[idx1, 0], vectors_2d[idx2, 0]], 
                [vectors_2d[idx1, 1], vectors_2d[idx2, 1]], 
                'r--', alpha=0.3, linewidth=1, label='ä½ç›¸ä¼¼åº¦' if i == 1 else '')

ax2.set_title('è¯­ä¹‰å‘é‡ç©ºé—´å¯è§†åŒ–ï¼ˆPCAåˆ°2Dï¼‰', fontsize=12, fontweight='bold')
ax2.set_xlabel('PC1')
ax2.set_ylabel('PC2')
ax2.legend()
ax2.grid(True, alpha=0.3)

# å·¦ä¸‹ï¼šç›¸ä¼¼åº¦çŸ©é˜µçƒ­åŠ›å›¾
ax3 = axes[1, 0]
all_unique_texts = list(set([t for pair in instruction_pairs for t in pair[:2]]))
sim_matrix = np.zeros((len(all_unique_texts), len(all_unique_texts)))
for i, text1 in enumerate(all_unique_texts):
    for j, text2 in enumerate(all_unique_texts):
        if i == j:
            sim_matrix[i, j] = 1.0
        elif text1 in vectors and text2 in vectors:
            sim_matrix[i, j] = np.dot(vectors[text1], vectors[text2])
        else:
            sim_matrix[i, j] = 0.0

im = ax3.imshow(sim_matrix, cmap='viridis', vmin=0, vmax=1, aspect='auto')
ax3.set_title('è¯­ä¹‰ç›¸ä¼¼åº¦çŸ©é˜µ', fontsize=12, fontweight='bold')
ax3.set_xlabel('æŒ‡ä»¤')
ax3.set_ylabel('æŒ‡ä»¤')
ax3.set_xticks(range(len(all_unique_texts)))
ax3.set_yticks(range(len(all_unique_texts)))
ax3.set_xticklabels(all_unique_texts, rotation=45, ha='right', fontsize=9)
ax3.set_yticklabels(all_unique_texts, fontsize=9)
for i in range(len(all_unique_texts)):
    for j in range(len(all_unique_texts)):
        ax3.text(j, i, f'{sim_matrix[i, j]:.2f}', ha='center', va='center', 
                fontsize=8, fontweight='bold', 
                color='white' if sim_matrix[i, j] < 0.5 else 'black')
plt.colorbar(im, ax=ax3, label='ç›¸ä¼¼åº¦')

# å³ä¸‹ï¼šç›¸ä¼¼åº¦åˆ†å¸ƒ
ax4 = axes[1, 1]
sim_values = [s for _, _, s in instruction_pairs]
ax4.hist(sim_values, bins=10, color='steelblue', alpha=0.7, edgecolor='black', linewidth=1.5)
ax4.set_title('è¯­ä¹‰ç›¸ä¼¼åº¦åˆ†å¸ƒ', fontsize=12, fontweight='bold')
ax4.set_xlabel('ç›¸ä¼¼åº¦')
ax4.set_ylabel('é¢‘æ•°')
ax4.grid(True, alpha=0.3, axis='y')

plt.tight_layout()
plt.suptitle('è¯­ä¹‰ç›¸ä¼¼åº¦å¯è§†åŒ–ï¼šVLAæŒ‡ä»¤çš„è¯­ä¹‰ç›¸ä¼¼åº¦', fontsize=14, fontweight='bold', y=1.0)
plt.show()

print("=" * 60)
print("è¯­ä¹‰ç›¸ä¼¼åº¦å¯è§†åŒ–è¯´æ˜ï¼š")
print("=" * 60)
print("1. å·¦ä¸Šï¼šè¯­ä¹‰ç›¸ä¼¼åº¦å¯¹æ¯”ï¼Œæ˜¾ç¤ºç›®æ ‡ç›¸ä¼¼åº¦å’Œå®é™…ç›¸ä¼¼åº¦")
print("2. å³ä¸Šï¼šè¯­ä¹‰å‘é‡ç©ºé—´å¯è§†åŒ–ï¼Œç›¸ä¼¼åº¦é«˜çš„æŒ‡ä»¤åœ¨ç©ºé—´ä¸­é è¿‘")
print("3. å·¦ä¸‹ï¼šè¯­ä¹‰ç›¸ä¼¼åº¦çŸ©é˜µï¼Œæ˜¾ç¤ºæ‰€æœ‰æŒ‡ä»¤å¯¹ä¹‹é—´çš„ç›¸ä¼¼åº¦")
print("4. å³ä¸‹ï¼šè¯­ä¹‰ç›¸ä¼¼åº¦åˆ†å¸ƒï¼Œæ˜¾ç¤ºç›¸ä¼¼åº¦çš„åˆ†å¸ƒæƒ…å†µ")
print("=" * 60)
```

### 3.3 è¯­ä¹‰è§’è‰²æ ‡æ³¨çš„æ•°å­¦è¡¨ç¤º

è¯­ä¹‰è§’è‰²æ ‡æ³¨çš„æ•°å­¦è¡¨ç¤ºä¸ºï¼š

å¯¹äºè¾“å…¥åºåˆ— $x = [x_1, x_2, \ldots, x_n]$ï¼Œè¯­ä¹‰è§’è‰²æ ‡æ³¨è¾“å‡ºæ ‡ç­¾åºåˆ— $y = [y_1, y_2, \ldots, y_n]$ï¼Œå…¶ä¸­ï¼š

$$y_i = \text{SRL}(\text{Encoder}(x), i)$$

å…¶ä¸­ï¼š
- $\text{Encoder}(x)$ æ˜¯è¯­è¨€ç¼–ç å™¨æå–çš„ç‰¹å¾
- $\text{SRL}(\cdot)$ æ˜¯è¯­ä¹‰è§’è‰²æ ‡æ³¨å±‚
- $y_i$ æ˜¯ç¬¬$i$ä¸ªè¯è¯­çš„è¯­ä¹‰è§’è‰²æ ‡ç­¾

### 3.4 è¯­ä¹‰è§’è‰²æ ‡æ³¨çš„å¯è§†åŒ–

ä¸‹é¢æˆ‘ä»¬é€šè¿‡ä»£ç å¯è§†åŒ–è¯­ä¹‰è§’è‰²æ ‡æ³¨çš„è¿‡ç¨‹ï¼š


```python
# ============================================
# è¯­ä¹‰è§’è‰²æ ‡æ³¨å¯è§†åŒ–ï¼ˆç¤ºä¾‹ï¼šVLAæŒ‡ä»¤çš„è¯­ä¹‰è§’è‰²æ ‡æ³¨ï¼‰
# ============================================
np.random.seed(42)

# ç¤ºä¾‹ï¼šVLAæŒ‡ä»¤
instruction = "æ‹¿èµ·æ¡Œå­ä¸Šçš„çº¢è‰²æ¯å­"
tokens = ["æ‹¿èµ·", "æ¡Œå­", "ä¸Š", "çš„", "çº¢è‰²", "æ¯å­"]
seq_len = len(tokens)

# è¯­ä¹‰è§’è‰²
semantic_roles = ['åŠ¨ä½œ', 'å—äº‹', 'åœ°ç‚¹', 'å±æ€§', 'O']
num_roles = len(semantic_roles)

# è¯­ä¹‰è§’è‰²æ ‡æ³¨ç»“æœ
role_labels = ["åŠ¨ä½œ", "åœ°ç‚¹", "åœ°ç‚¹", "O", "å±æ€§", "å—äº‹"]

# æ¨¡æ‹Ÿæ¨¡å‹é¢„æµ‹çš„æ¦‚ç‡ï¼ˆç®€åŒ–ç‰ˆï¼‰
np.random.seed(42)
role_probs = np.random.rand(seq_len, num_roles)
role_probs = role_probs / role_probs.sum(axis=1, keepdims=True)

# è®¾ç½®åˆç†çš„æ¦‚ç‡ï¼ˆæ¨¡æ‹Ÿæ¨¡å‹é¢„æµ‹ï¼‰
for i, (token, label) in enumerate(zip(tokens, role_labels)):
    if label in semantic_roles:
        role_idx = semantic_roles.index(label)
        role_probs[i, role_idx] = 0.8
        role_probs[i, :] = role_probs[i, :] / role_probs[i, :].sum()

# å¯è§†åŒ–
fig, axes = plt.subplots(2, 1, figsize=(16, 10))

# ä¸Šå›¾ï¼šè¯­ä¹‰è§’è‰²æ ‡æ³¨å¯è§†åŒ–
ax1 = axes[0]
colors_map = {
    'åŠ¨ä½œ': 'red',
    'å—äº‹': 'blue',
    'åœ°ç‚¹': 'green',
    'å±æ€§': 'orange',
    'O': 'gray'
}

y_pos = np.arange(seq_len)
for i, (token, label) in enumerate(zip(tokens, role_labels)):
    color = colors_map.get(label, 'gray')
    ax1.barh(i, 1, color=color, alpha=0.7, edgecolor='black', linewidth=1.5)
    ax1.text(0.5, i, f'{token} ({label})', ha='center', va='center', 
            fontsize=11, fontweight='bold')

ax1.set_yticks(y_pos)
ax1.set_yticklabels(tokens)
ax1.set_xlabel('ä½ç½®')
ax1.set_ylabel('è¯è¯­')
ax1.set_title('è¯­ä¹‰è§’è‰²æ ‡æ³¨å¯è§†åŒ–ï¼šVLAæŒ‡ä»¤çš„è¯­ä¹‰è§’è‰²æ ‡æ³¨', fontsize=12, fontweight='bold')
ax1.set_xlim(0, 1)
ax1.invert_yaxis()
ax1.grid(True, alpha=0.3, axis='x')

# ä¸‹å›¾ï¼šé¢„æµ‹æ¦‚ç‡å¯è§†åŒ–
ax2 = axes[1]
x_pos = np.arange(seq_len)
width = 0.15

for i, role in enumerate(semantic_roles):
    probs = role_probs[:, i]
    offset = (i - num_roles/2) * width
    ax2.bar(x_pos + offset, probs, width, label=role, alpha=0.7, 
           edgecolor='black', linewidth=1)

ax2.set_xlabel('è¯è¯­ä½ç½®')
ax2.set_ylabel('æ¦‚ç‡')
ax2.set_title('é¢„æµ‹æ¦‚ç‡åˆ†å¸ƒ', fontsize=12, fontweight='bold')
ax2.set_xticks(x_pos)
ax2.set_xticklabels(tokens, rotation=45, ha='right')
ax2.set_ylim(0, 1.1)
ax2.legend()
ax2.grid(True, alpha=0.3, axis='y')

plt.tight_layout()
plt.suptitle('è¯­ä¹‰è§’è‰²æ ‡æ³¨å¯è§†åŒ–ï¼šVLAæŒ‡ä»¤çš„è¯­ä¹‰è§’è‰²æ ‡æ³¨', fontsize=14, fontweight='bold', y=1.0)
plt.show()

print("=" * 60)
print("è¯­ä¹‰è§’è‰²æ ‡æ³¨å¯è§†åŒ–è¯´æ˜ï¼š")
print("=" * 60)
print("1. ä¸Šå›¾ï¼šè¯­ä¹‰è§’è‰²æ ‡æ³¨å¯è§†åŒ–ï¼Œæ˜¾ç¤ºæ¯ä¸ªè¯è¯­çš„è¯­ä¹‰è§’è‰²")
print("2. ä¸‹å›¾ï¼šé¢„æµ‹æ¦‚ç‡åˆ†å¸ƒï¼Œæ˜¾ç¤ºæ¨¡å‹å¯¹æ¯ä¸ªè¯è¯­çš„è¯­ä¹‰è§’è‰²é¢„æµ‹æ¦‚ç‡")
print("3. è¯­ä¹‰è§’è‰²æ ‡æ³¨çš„ç›®æ ‡æ˜¯è¯†åˆ«æŒ‡ä»¤ä¸­è¯è¯­çš„è¯­ä¹‰è§’è‰²")
print("=" * 60)
```
